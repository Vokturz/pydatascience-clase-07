```python
import pandas as pd

df = pd.read_csv('../data/universal_top_spotify_songs.csv')
df['date'] = pd.to_datetime(df['snapshot_date'])
df.head()
```

# Descripción de columnas

- `spotify_id`: El ID de la canción es spotify
- `name`: El nombre de la canción
- ...

```python
df.info()
```

```python
df_chile = df[df['country'] == 'CL']
df_chile.head(10)
```

```python
artista_df = df_chile[(df_chile['name'] == 'SI NO ES CONTIGO') & (df_chile['artists'] == 'Cris Mj')]
artista_df2 = df_chile[(df_chile['name'] == 'No Ponga Excusas') & (df_chile['artists'] == 'Cris Mj')]
ax = artista_df.plot('date', 'daily_rank')
artista_df2.plot('date', 'daily_rank', ax=ax)
```

## Cuantas veces aparece "Cris Mj" como artista en el top 50 de Chile?

```python
crismj_df = df_chile[['artists', 'date']].dropna()
crismj_df['aparece_mj'] = crismj_df['artists'].apply(lambda x : 'Cris Mj' in x)
crismj_df
```

```python
crismj_df.groupby('date')['aparece_mj'].sum().plot()
```
